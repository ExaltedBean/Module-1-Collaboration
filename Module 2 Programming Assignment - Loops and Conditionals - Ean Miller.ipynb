```python
secret = 4
guess = 6
if guess > secret:
    print("Too High")
elif guess < secret:
    print("Too Low")
elif guess == secret:
    print("Just Right")
```

```python
small = False
green = False
if green == True and small == True:
    print("It is a pea.")
elif green == False and small == True:
    print("It is a cherry")
elif green == False and small == False:
    print("It is a pumpkin")
elif green == True and small == False:
    print("It is a watermelon")
    
```

```python
guess_me = 7
number = 1
while True:
    if number < guess_me:
        print("Too Low")
    elif number == guess_me:
        print("Found it!")
        break
    elif number > guess_me:
        print("Oops!")
        break
    number += 1
```

```python
guess_me = 5
for number in range(10):
    if number < guess_me:
        print("Too Low")
    elif number == guess_me:
        print("Found it!")
        break
    elif number > guess_me:
        print("Oops!")
        break
    number += 1
```
