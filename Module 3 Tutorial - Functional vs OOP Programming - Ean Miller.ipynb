```python
class Solution:
    def sort012(self, arr):
        """
        Sorts an array containing only 0s, 1s, and 2s.
        No built-in sorting allowed as the assignment states, so we follow a common method that works around it.
        
        Essentially what this does makes:
        - 0s get pushed to the front
        - 2s get pushed to the back
        - 1s naturally settle in the middle
        - All of this to ensure that the array is properly sorted.
        """

        low, mid, high = 0, 0, len(arr) - 1

        # Keep going until mid crosses high
        while mid <= high:
            # If we see a 0, swap it to the front
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1

            # If it's a 1, it's already in the right zone
            elif arr[mid] == 1:
                mid += 1

            # If it's a 2, swap it to the back
            else:
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

        return arr


# Example usage
arr = [0, 1, 2, 0, 1, 2]
Solution().sort012(arr)
```

```python
def search_demo(arr, target):
    """
    Demonstrates two different ways: Binary Search and Lambda Search
    Both results are printed so you can compare the two approaches.
    """

    print("Array:", arr)
    print("Target:", target)

    # Lambda Search
    # -----------------------------
    # Filter keeps only elements equal to the target.
    # any() checks if there is at least one match.
    # this is only an attempt, my previous classes didn't cover Lambda.
    lambda_found = any(filter(lambda x: x == target, arr))
    print("Lambda Search Found:", lambda_found)
    
    # Binary Search
    # -----------------------------
    low, high = 0, len(arr) - 1
    binary_index = -1  # default if not found

    # Standard binary search loop which also makes sure to name the correct index position.
    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            binary_index = mid
            break
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    print("Binary Search Index:", binary_index)


# Example usage
arr = [1, 3, 5, 7, 9, 11, 13]
target = 7
search_demo(arr, target)
```
